#preprocessing the merged data for model buiding(Basic Neural Network)

```python
import nltk
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
merged_df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#it is simpler option to take first k vocabularies and simpler to computation but most of the important words will be misse
tokenizer = Tokenizer( oov_token='<OOV>')
tokenizer.fit_on_texts(merged_df['Transcription'])
sequences = tokenizer.texts_to_sequences(merged_df['Transcription'])
padded_sequences = pad_sequences(sequences, maxlen=Max_count, padding='post', truncating='post')
    
    
```

```python
padded_sequences,padded_sequences.shape
```

#Simple forward neural network

```python
from keras.backend import dropout
import tensorflow as tf

# Create a sequential model
model = tf.keras.Sequential()

# Add input layer
model.add(tf.keras.layers.Input(shape=(163,)))

# Add fully connected hidden layers with ReLU activation
model.add(tf.keras.layers.Dense(128, activation='tanh'))
model.add(tf.keras.layers.Dropout(0.2))

model.add(tf.keras.layers.Dense(64, activation='tanh'))

model.add(tf.keras.layers.Dense(32, activation='tanh'))
model.add(tf.keras.layers.Dropout(0.3))
# Add output layer with sigmoid activation
model.add(tf.keras.layers.Dense(1, activation='tanh'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
labels=df.Veracity.to_list()
labels=np.array(labels)
accuracy_tr=np.array(df.Accuracy)
anxious=np.array(df.Anxious)
valence=np.array(df.Valence)
accuracy_tr=accuracy_tr.reshape(320,1)
anxious=anxious.reshape(320,1)
valence=valence.reshape(320,1)
```

```python
concatenated_data = np.concatenate([padded_sequences,accuracy_tr, anxious,valence], axis=1)
concatenated_data.shape,padded_sequences.shape
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(concatenated_data,labels,test_size=0.2,random_state=0)
```

```python
X_train.shape,y_train.shape
```

```python
model.fit(X_train,y_train,epochs=20)
```

```python
test_loss, test_acc = model.evaluate(X_test,y_test)
```

```python
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 100
max_seq_length = 163
```

#RNN MODEL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout

# define the model architecture
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_seq_length))
model.add(LSTM(units=64, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(units=1, activation='tanh'))

# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# train the model
model.fit(X_train, y_train, batch_size=16, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(X_test,y_test)
```
